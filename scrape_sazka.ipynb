```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.sazka.cz/kurzove-sazky/sports/competition/8183/tenis/atp-tour/atp-kitzbuhel-dvouhra/matches'  # Replace this with the URL of the website you want to scrape

# Send an HTTP request to the URL and get the HTML content
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    print(soup)

    # Find the div with the specified class and data-testid attributes
    div_element = soup.find('div', class_='eventCardTeamName-0-3-993 eventCardTeamName-d0-0-3-994')

    # Check if the div element was found
    if div_element:
        # Get the text inside the div (the content)
        content = div_element.text
        print(content)
    else:
        print("Div element not found.")
else:
    print("Failed to fetch the URL:", response.status_code)
```
