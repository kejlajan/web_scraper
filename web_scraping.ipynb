### useful functions:

Function to 'grep' something and return the row of a DF:

```python
def GrepRow(string, DF):
    found = False
    for index in DF.index:
        for column in DF.columns:
            if (str(DF.loc[index,column]) in str(string))  or  (str(string) in str(DF.loc[index,column])):
                print(DF.iloc[index].to_string())
                found = True
                break
    if found == False:
        print("Nothing found...")
        
```

### Open files and store as (multi-level) dictionaries:

```python
with open(from_chance_json) as f:
    odds_chance_dict = json.load(f)
```

```python
for index, row in enumerate(odds_tipsport_dict["offerSuperSports"][0]["tabs"][0]["offerCompetitionAnnuals"][1]["matches"][0]["oppRows"][0]["oppsTab"]):
        print(f"*****{index} \n {row}")
```

```python
opportunities = []

for index, row in enumerate(odds_tipsport_dict["offerSuperSports"][0]["tabs"][0]["offerCompetitionAnnuals"]):
    if row['matches'][0]['homeParticipant']!=None:
        opportunities.append({"host": row['matches'][0]["homeParticipant"] , "guest" : row['matches'][0]["visitingParticipant"], "host_odd" : row["matches"][0]["oppRows"][0]["oppsTab"][0]["odd"], "guest_odd" : row["matches"][0]["oppRows"][0]["oppsTab"][4]["odd"]})
```

```python
print(opportunities)
```

```python
tipsport_opportunities_DF = pd.DataFrame(opportunities)
tipsport_opportunities_DF
```

```python
tipsport_opportunities_DF["theor_likelihood"] = round( ((1/tipsport_opportunities_DF["host_odd"])  +  (1/tipsport_opportunities_DF["guest_odd"])), 3)
tipsport_opportunities_DF
```

```python
tipsport_opportunities_DF.sort_values(by="theor_likelihood")
```
